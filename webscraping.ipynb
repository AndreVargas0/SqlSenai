```python
import pandas as pd

url = 'https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/#google_vignette'

tabela = pd.read_html(url, encoding='utf-8')

type(tabela)
```

```python
display(tabela)
```

```python
# Passo 1.10: Mostar a converversão de lista para dataframe

tabela_brasileirao = tabela[0]

display(tabela_brasileirao)
```

```python
df = tabela_brasileirao

# Passo 1.11: Mostar os cinco primeiros 

df.head()
```

```python
# Passo 1.12: mostar os cinco últimos

df.tail()
```

```python
### Passo 1.13: Mostar todas as colunas

df.columns
```

```python
### Passo 1.14: Mostar uma coluna

df['Times.2']
```

```python
### Passo 1.15: Mostar varias colunas

df[['Times.2','V','D']]
```

```python
### Passo 1.16: Mostar informações estruturarias do dataframe

df.info()
```

```python
### Passo 1.17: Mostar informações de estatística descritiva do dataframe

df.describe()
```

```python
### Passo 1.18: Mostar informações da coluna e %

df[['Times.2','%']]
```

```python
### Passo 1.19: Renomear nome de uma coluna

df.rename(columns={'Times.2':'times'},inplace=True)
df.rename(columns={'%':'percentual'},inplace=True)
```

```python
df[['times','percentual']]
```
